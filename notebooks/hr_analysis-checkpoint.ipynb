```python
import pandas as pd
import matplotlib.pyplot as plt

# Load cleaned data
df = pd.read_csv("cleaned_hr_data.csv")

# Convert Attrition to binary
df['attrition_flag'] = df['attrition'].map({'Yes': 1, 'No': 0})

# Attrition by Age
df.groupby('age')['attrition_flag'].mean().plot(title="Attrition by Age")
plt.show()

# Attrition by Salary
df.groupby('monthlyincome')['attrition_flag'].mean().plot(title="Attrition by Salary")
plt.show()

# Attrition by Experience
df.groupby('totalworkingyears')['attrition_flag'].mean().plot(title="Attrition by Experience")
plt.show()

# Satisfaction vs Attrition
df.groupby('jobsatisfaction')['attrition_flag'].mean().plot(title="Satisfaction vs Attrition")
plt.show()
```
